```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("student_data.csv")
```

```python
def label_risk(row):
    if row["Attendance_Percentage"] < 60 or row["Internal_Marks"] < 50:
        return 1  # At Risk
    else:
        return 0  # Not At Risk

data["Dropout_Risk"] = data.apply(label_risk, axis=1)
```

```python
X = data[["Attendance_Percentage", "Internal_Marks", "Assignments_Submitted"]]
y = data["Dropout_Risk"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
data["Predicted_Risk"] = model.predict(X)

data["Risk_Level"] = data["Predicted_Risk"].apply(lambda x: "High Risk" if x == 1 else "Low Risk")

data[["Student_ID", "Risk_Level"]]
```
